```python
features = df.drop(['country', 'year','population','cont','life_exp'], axis=1)
target= df.life_exp.values
```

```python
import numpy as np
logged_target = np.log(target)
sns.histplot(logged_target)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(features, logged_target, test_size=0.20, random_state=42)
```

# LinearRegression Model

```python
from sklearn.linear_model import LinearRegression, SGDRegressor
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predicted= model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

mae = mean_absolute_error(y_test, predicted)
mse = mean_squared_error(y_test, predicted)
r2s = r2_score(y_test, predicted)
print('mae= ', mae)
print('mse= ', mse)
print('r2s= ', r2s)
```

```python
comp = np.column_stack((np.exp(y_test), np.exp(predicted)))
print(comp)
```

```python
sns.regplot(x=comp[:,0], y=comp[:,1])
```

# DecisionTreeRegressor

```python
features = df.drop(['country','year','life_exp'],axis=1)
target=df.life_exp.values
```

```python
nums_features = features.select_dtypes('number')
text_features=features.select_dtypes('object')
```

```python
text_features
```

```python
from  sklearn.preprocessing  import OneHotEncoder

ohe=OneHotEncoder(sparse=False)
ohe.fit(text_features)
ohe_data=  ohe.transform(text_features)
ohe_data=pd.DataFrame(ohe_data,columns=ohe.get_feature_names_out())
```

```python
ohe_data
```

```python
full_data=pd.concat([ohe_data,nums_features],axis=1)
```

```python
full_data
```

```python
full_data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(full_data, logged_target, test_size=0.20, random_state=42)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score 

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(X_train,y_train)
pre= dt.predict(X_test)
mae = mean_absolute_error(y_test, pre)
mse = mean_squared_error(y_test, pre)
r2s = r2_score(y_test, pre)
print('mae= ', mae)
print('mse= ', mse)
print('r2s= ', r2s)
```

```python
predict=dt.predict(X_test)
comp=np.column_stack((np.exp(y_test),np.exp(predict)))
comp
```

```python
sns.regplot(x=comp[:,0], y=comp[:,1])
```
